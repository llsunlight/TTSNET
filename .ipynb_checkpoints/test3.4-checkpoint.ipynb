```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat, savemat
from tqdm import tqdm
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from joblib import Parallel, delayed
from scripts.dataset import Dataset
from scripts.model import STRCA
from scipy.linalg import eigh 

if __name__ == "__main__":
    Fs = 256
    Co_range = np.arange(1, 12)
    num_fold = 5
    num_subj = 15
    max_freq = 10
    step_bank = 2
    frange = np.arange(step_bank, max_freq + 1, step_bank)
    channels = (np.array([1, 3, 5, 15, 29, 30, 31, 45, 55, 57, 59]) - 1).tolist()
    num_bank = int(max_freq / step_bank)
    t = [0, 2]
    Co_performance = np.zeros([num_subj, num_fold, len(frange), len(Co_range)])
    results = np.zeros([num_subj, num_fold, len(Co_range)])  
    best_Co = np.zeros([num_subj, num_fold, len(frange)], dtype=int)

    for sub in tqdm(range(num_subj), desc='Processing Subjects'):
        for fold in range(num_fold):
            dataset = Dataset(num_class=7, path='Dataset/OData/Dataset I/', channel=channels)
            dataset.load_data(sub, 'zscore')
            dataset.divide_data(num_fold)
            dataset.filter_data(fs=Fs, frange=frange, order=8, btype='lowpass')
            
            for f, freq in enumerate(frange):
                train_set, train_label, _, _ = dataset.retreive_data(
                    class_index=list(range(7)), fold_index=fold, filter_index=f+1)
                train_set = train_set[:, :, int(t[0] * Fs):int(t[1] * Fs)]
                max_score = 0
                max_co = 1

                for Co in Co_range:
                    model = Pipeline([
                        ('strca', STRCA(Co, optimizer='trca', ifG=True)),
                        ('clf', SVC())
                    ])
                    model.fit(train_set, train_label)
                    score = model.score(train_set, train_label)
                    if score > max_score:
                        max_score = score
                        max_co = Co
                    results[sub, fold, Co - 1] = score  
                    Co_performance[sub, fold, f, Co - 1] = score  
                best_Co[sub, fold, f] = max_co  

    savemat('gridcv_co_trca_test4_11.mat', {'Co_performance': Co_performance})
    savemat('gridcv_results_test4_11.mat', {'results': results})
    savemat('best_Co.mat', {'best_Co': best_Co})
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scripts.dataset import Dataset
from scripts.model import STRCA
import random

random_seed = 74
np.random.seed(random_seed)
random.seed(random_seed)

def plot_sorted_features(filtered_data, Co, offset=10.0, Fs=256):
    # 确保输入数据是三维的
    if filtered_data.ndim != 3:
        raise ValueError(f"Expected 3D input for filtered_data, but got {filtered_data.ndim}D")

    # 计算每个通道的方差
    feature_variances = np.var(filtered_data, axis=2)  # 对时间维度计算方差
    sorted_indices = np.argsort(feature_variances, axis=1)[:, ::-1]  # 对通道排序，而不是时间
    
    # 对每个 trial 的通道进行排序
    sorted_data = np.empty_like(filtered_data)
    for i in range(filtered_data.shape[0]):
        sorted_data[i, :, :] = filtered_data[i, sorted_indices[i, :], :]

    plt.figure(figsize=(8, 6))
    
    num_features = sorted_data.shape[1]
    time_axis = np.arange(sorted_data.shape[2]) / Fs  # 基于时间点数生成时间轴
    
    for i in range(num_features):
        plt.plot(time_axis, sorted_data[0, i, :] + i * offset, label=f'Feature {i+1}')
    
    plt.title(f'Filtered EEG Signals (Sorted by Feature Variance, Co={Co})')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (with offset)')
    plt.grid(True)
    plt.show()

def apply_spatial_filtering(data, W):
    """
    Apply spatial filter W to the data.
    Args:
    - data: Input EEG data of shape (trials, channels, time)
    - W: Spatial filter matrix of shape (channels, channels)
    
    Returns:
    - filtered_data: Spatially filtered data of shape (trials, channels, time)
    """
    return np.einsum('ij,kjl->kil', W, data)

def process_and_plot(Fs, sub, channels, Co):
    t = [0, 2]

    dataset = Dataset(num_class=7, path='Dataset/OData/Dataset I/', channel=channels)
    dataset.load_data(sub, None)
    dataset.divide_data(5)
    dataset.filter_data(fs=Fs, frange=[6, 8], order=8, btype='lowpass')

    train_set, train_label, test_set, test_label = dataset.retreive_data(
        class_index=[0,1,2,3,4,5,6], fold_index=1, filter_index=[1])

    # 将数据切割成时间窗口
    train_set = train_set[:,:,int(t[0]*Fs):int(t[1]*Fs),:]
    test_set = test_set[:,:,int(t[0]*Fs):int(t[1]*Fs),:]
    
    # 去掉多余的维度 (trials, channels, time, 1) -> (trials, channels, time)
    if train_set.ndim == 4:
        train_set = np.squeeze(train_set, axis=-1)
    if test_set.ndim == 4:
        test_set = np.squeeze(test_set, axis=-1)
    
    trca = STRCA(Co=Co, optimizer='trca', ifG=True)
    trca.fit(train_set, train_label)
    
    # 获取空间滤波器 W
    W = trca.W
    
    # 手动应用空间滤波器
    filtered_test_set = apply_spatial_filtering(test_set[test_label == 1], W)
    
    # 确保 filtered_test_set 始终是三维的
    if filtered_test_set.ndim != 3:
        raise ValueError(f"Expected 3D filtered_test_set, but got {filtered_test_set.ndim}D")
    
    plot_sorted_features(filtered_test_set, Co, Fs=Fs)

if __name__ == "__main__":
    Fs = 256
    sub = 10
    channels = (np.int64([1, 3, 5, 15, 29, 30, 31, 45, 55, 57, 59]) - 1).tolist()
    Co = 11
    
    process_and_plot(Fs, sub, channels, Co)
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from scripts.dataset import Dataset
from scripts.model import STRCA, FBTRCA
from tqdm import tqdm
from joblib import Parallel, delayed
from scipy.io import savemat, loadmat
import os

if __name__ == "__main__":
    Fs = 256
    Co = 11
    num_fold = 5
    num_subj = 15
    max_freq = 10
    step_bank = 2
    frange = np.arange(step_bank, max_freq + 1, step_bank)
    channels = (np.int64([1, 3, 5, 15, 29, 30, 31, 45, 55, 57, 59]) - 1).tolist()
    t = [0, 2]
    result = np.zeros([num_subj, num_fold, 4])
    
    gridcv_Co_trca = np.zeros([num_subj, num_fold, len(frange)], dtype=np.int64)
    gridcv_Co_tdlda = np.zeros([num_subj, num_fold, len(frange)], dtype=np.int64)

    if not (os.path.exists('gridcv_co_trca_31.mat') and os.path.exists('gridcv_co_tdlda_31.mat')):
        for sub in range(num_subj):
            for fold in tqdm(range(num_fold), desc=f'sub{sub}:'):
                dataset = Dataset(num_class=7, path='Dataset/OData/Dataset I/', channel=channels)
                dataset.load_data(sub, 'zscore')
                dataset.divide_data(num_fold)
                dataset.filter_data(fs=Fs, frange=frange, order=8, btype='lowpass')
                
                def lambda_x(f, alg):
                    # Set the parameters for GridSearchCV
                    parameters = {'model__Co': np.arange(Co) + 1}
                    
                    # Retrieve the training data
                    train_set, train_label, _, _ = dataset.retreive_data(
                        class_index=[0,1,2,3,4,5,6], fold_index=fold, filter_index=f+1)
                    train_set = train_set[:,:,int(t[0]*Fs):int(t[1]*Fs)]
                    
                    # Initialize the model based on the algorithm type
                    if alg == 'trca':
                        model = STRCA(Co, optimizer='trca', ifG=True)
                    else:  # tdlda
                        model = STRCA(Co, optimizer='tdlda', ifG=True)
                    
                    # Create a pipeline with the model and SVM classifier
                    pipeline = Pipeline([('model', model), ('clf', SVC())])
                    
                    # Run GridSearchCV
                    clf = GridSearchCV(pipeline, parameters, scoring="accuracy", cv=4)
                    clf.fit(train_set, train_label)
                    return clf.best_params_['model__Co']
                
                # Perform parallel processing for TRCA and TD-LDA
                gridcv_Co_trca[sub, fold, :] = Parallel(n_jobs=len(frange))(delayed(lambda_x)(f, 'trca') for f in range(len(frange)))
                gridcv_Co_tdlda[sub, fold, :] = Parallel(n_jobs=len(frange))(delayed(lambda_x)(f, 'tdlda') for f in range(len(frange)))
        
        # Save the results to .mat files
        savemat('gridcv_co_trca_New11.mat', {'gridcv_Co_trca': gridcv_Co_trca})
        savemat('gridcv_co_tdlda_New11.mat', {'gridcv_Co_tdlda': gridcv_Co_tdlda})
    else:
        # Load the results if the files already exist
        gridcv_Co_trca = loadmat('gridcv_co_trca_New11.mat')['gridcv_Co_trca']
        gridcv_Co_tdlda = loadmat('gridcv_co_tdlda_New11.mat')['gridcv_Co_tdlda']
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scripts.dataset import Dataset
from scripts.model import STRCA
import random

random_seed = 1
np.random.seed(random_seed)
random.seed(random_seed)

def plot_sorted_features(grand_average_data, Co, offset=15.0, Fs=256):
    # 确保输入数据是三维的
    if grand_average_data.ndim != 2:
        raise ValueError(f"Expected 2D input for grand_average_data, but got {grand_average_data.ndim}D")

    # 计算每个通道的方差
    feature_variances = np.var(grand_average_data, axis=1)  # 对时间维度计算方差
    sorted_indices = np.argsort(feature_variances)  # 对通道从小到大排序

    # 对通道进行排序
    sorted_data = grand_average_data[sorted_indices, :]

    plt.figure(figsize=(8, 6))
    
    num_features = sorted_data.shape[0]
    time_axis = np.arange(sorted_data.shape[1]) / Fs  # 基于时间点数生成时间轴
    
    for i in range(num_features):
        plt.plot(time_axis, sorted_data[i, :] + i * offset)
    
    plt.title(f'')
    plt.xlabel('Time (s)')
    plt.ylabel('')
    plt.grid(True)
    plt.legend()
    plt.show()

def apply_spatial_filtering(data, W):
    return np.einsum('ij,kjl->kil', W, data)

def process_and_plot(Fs, sub, channels, Co):
    t = [0, 2]

    dataset = Dataset(num_class=7, path='Dataset/OData/Dataset I/', channel=channels)
    dataset.load_data(sub, None)
    dataset.divide_data(5)
    dataset.filter_data(fs=Fs, frange=[6, 8], order=8, btype='lowpass')

    train_set, train_label, test_set, test_label = dataset.retreive_data(
        class_index=[0,1,2,3,4,5,6], fold_index=1, filter_index=[1])

    train_set = train_set[:,:,int(t[0]*Fs):int(t[1]*Fs),:]
    test_set = test_set[:,:,int(t[0]*Fs):int(t[1]*Fs),:]
    
    if train_set.ndim == 4:
        train_set = np.squeeze(train_set, axis=-1)
    if test_set.ndim == 4:
        test_set = np.squeeze(test_set, axis=-1)
    
    trca = STRCA(Co=Co, optimizer='trca', ifG=True)
    trca.fit(train_set, train_label)
    
    W = trca.W
    
    # 获取对应类标签的测试数据
    filtered_test_set = apply_spatial_filtering(test_set[test_label == 1], W)
    
    # 计算 Grand Average
    if filtered_test_set.ndim != 3:
        raise ValueError(f"Expected 3D filtered_test_set, but got {filtered_test_set.ndim}D")
    
    grand_average_data = np.mean(filtered_test_set, axis=0)  # 在试验维度上取平均
    plot_sorted_features(grand_average_data, Co, Fs=Fs)

if __name__ == "__main__":
    Fs = 256
    sub = 8
    channels = (np.int64([1,3,5,12,15,18,20,21,24,25,29,30,31,35,36,39,40,45,55,57,59]) - 1).tolist()
    Co = 21
    
    process_and_plot(Fs, sub, channels, Co)
```
